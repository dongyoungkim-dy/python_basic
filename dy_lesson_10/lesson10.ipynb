## Data manipulation

```python
# Fill up NAs with values
train_data.fillna("This is a NULL Value")
```

```python
clean_data = train_data.copy()
clean_data["Cabin"] = clean_data["Cabin"].fillna("Just this column")
clean_data
```

```python
# Replace all "female" with "f"
clean_data = train_data.copy()
clean_data = clean_data.replace("female", "f")
clean_data
```

```python
# Replace all "female" with "f"
clean_data = train_data.copy()
# clean_data = clean_data.replace("female", "f")
clean_data.loc[clean_data["Sex"] == "female","Sex"] = "f"
clean_data
```

```python
# Impute missing ages with mean
clean_data = train_data.copy()
clean_data["Age"] = clean_data["Age"].fillna(clean_data["Age"].mean())
clean_data
```

```python
# Split the names
clean_data = train_data.copy()
clean_data["Name"] = clean_data["Name"].str.split()
clean_data
```

```python
# Split the names
clean_data = train_data.copy()
clean_data["Name"] = clean_data["Name"].str.strip(")")
clean_data
```

```python
# Split the names
clean_data = train_data.copy()
clean_data = clean_data.loc[clean_data["Name"].str.contains("\("), ]

clean_data["Name"] = clean_data["Name"].apply(lambda x : x[x.index("(") + 1:x.index(")")])
clean_data
```
