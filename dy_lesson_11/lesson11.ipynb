# Lesson 11 - Groupby and Aggregate

```python
import numpy as np
import pandas as pd
data = pd.read_csv("titanic_data/train.csv")
data
```

```python
data.groupby("Sex").count()
```

```python
for group in data.groupby("Sex"):
    print(group)
```

## Steps
1. Select which columns to group by with .groupby() function
2. Apply an aggregating function for each group

```python
# a format that I like to use
target_cols = ["PassengerId", "Survived"]
groupby_cols = ["Sex"]
all_cols = target_cols + groupby_cols

data[all_cols].groupby(groupby_cols).count()
```

```python
data[all_cols].groupby(groupby_cols).sum()
```

```python
data[all_cols].groupby(groupby_cols).agg({
    "PassengerId" : "count",
    "Survived" : "sum"
    })
```

```python
gender_survivde = data[all_cols].groupby(groupby_cols).agg({
    "PassengerId" : "count",
    "Survived" : "sum"
    })
gender_survivde["PassengerId"]
gender_survivde["Survival rate (%)"] = gender_survivde["Survived"]/gender_survivde["PassengerId"]*100
```

```python
gender_survivde = data[all_cols].groupby(groupby_cols).agg({
    "PassengerId" : "count",
    "Survived" : "mean"
    })
gender_survivde
```

```python
data_perc = data.copy()
target_cols = ["Survived"]
groupby_cols = ["Sex"]
all_cols = target_cols + groupby_cols
# Attach the percentage of survival of group of Sex to each individual
def percent(x):
    return 100*np.sum(x) / len(x)
data_perc["Percentage"] = data[all_cols].groupby(groupby_cols).transform(lambda x : 100*np.sum(x)/len(x))
data_perc
```

```python
target_cols = ["PassengerId"]
groupby_cols = ["Sex", "Pclass"]
all_cols = target_cols + groupby_cols

data[all_cols].groupby(groupby_cols).count()
```

```python
target_cols = ["PassengerId"]
groupby_cols = ["Pclass", "Sex"]
all_cols = target_cols + groupby_cols

groupby_result = data[all_cols].groupby(groupby_cols).count()
groupby_result
```

```python
target_cols = ["PassengerId", "Survived"]
groupby_cols = ["Pclass", "Sex"]
all_cols = target_cols + groupby_cols

groupby_result = data[all_cols].groupby(groupby_cols).agg({
    "PassengerId": "count",
    "Survived": "sum"
})
groupby_result["Percentage"] = round(groupby_result["Survived"]/groupby_result["PassengerId"]*100)
groupby_result
```

```python
target_cols = ["PassengerId", "Survived"]
groupby_cols = ["Pclass", "Sex"]
all_cols = target_cols + groupby_cols

groupby_result = data[all_cols].groupby(groupby_cols).agg({
    "PassengerId": "count",
    "Survived": ["sum","mean"]
})
groupby_result
```
